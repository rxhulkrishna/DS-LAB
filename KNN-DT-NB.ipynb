```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {'Color':np.random.choice(['Black','White','Red','Silver'],100),
        'Model':np.random.choice(['Hero','Kawasaki','BMW','TVS','KTM'],100),
        'Year':np.random.randint(2018,2023,100),
        'KM_Run':np.random.randint(50000,75000,100),
        'Buy':np.random.choice(['Yes','No'],100)}

df = pd.DataFrame(data)
df['Color'] = df['Color'].map({'Black':1,'White':2,'Red':3,'Silver':4})
df['Model'] = df['Model'].map({'Hero':0,'Kawasaki':1,'BMW':2,'TVS':3,'KTM':4})

x = df[['Color','Model','Year','KM_Run']]
y = df['Buy']

x_train,x_test,y_train,y_test = train_test_split(x.values,y,train_size=0.7,random_state=42)
```

```python
#KNN
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)

y_pred = knn.predict(x_test)
accuracy = accuracy_score(y_test,y_pred) * 100
print(f"Accuracy Score {accuracy}")

features = [[int(i) for i in input(f"Enter {x.columns} : ").split()]]
result = knn.predict(features)

print(f"Result : {result}")
```

```python
#DECISION TREE
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)

plt.figure(figsize=(18, 8))

plot_tree(dt)
plt.show()

y_pred = dt.predict(x_test)
accuracy = accuracy_score(y_test,y_pred) * 100
print(f"Accuracy Score : {accuracy}")

features = [[int(i) for i in input(f"Enter {x.columns} : ").split()]]
result = dt.predict(features)

print(f"Result : {result}")
```

```python
#NAIVE BAYES
nb = GaussianNB()
nb.fit(x_train,y_train)

y_pred = nb.predict(x_test)
accuracy = accuracy_score(y_test,y_pred) * 100
print(f"Accuracy Score : {accuracy}")

features = [[int(i) for i in input(f"Enter {x.columns} : ").split()]]
result = dt.predict(features)

print(f"Result : {result}")
```
