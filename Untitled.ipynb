```python
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
```

```python
data = pd.read_csv('diabetes.csv')
print(data)

x = data[['BMI','Age']]
y = data['Sugar']

data.plot.scatter('BMI','Age',c='Sugar',cmap='viridis')

clf = KNeighborsClassifier(n_neighbors=3)
clf.fit(x.values,y.values)

features = [[float(i) for i in input(f"Enter {x.columns} : ").split()]]
predict = clf.predict(features)

if(predict == 1):
    print("Diabetic")
else:
    print("Not Diabetic")
```

```python
from sklearn.datasets import load_iris
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
iris = load_iris()
x,y = iris.data,iris.target

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.7,random_state=1)
clf = tree.DecisionTreeClassifier()

clf.fit(x_train,y_train)
tree.plot_tree(clf)

predict = clf.predict(x_test)
accuracy = accuracy_score(y_test,predict) * 100
print(f"Accuracy : {accuracy}")

features = [[float(i) for i in input(f"Enter {iris.feature_names} : ").split()]]
result = clf.predict(features)

print(iris.target_names[result])
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
iris = load_iris()
x,y = iris.data,iris.target

x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.7)
clf = GaussianNB()

clf.fit(x_train,y_train)
predict = clf.predict(x_test)

accuracy = accuracy_score(y_test,predict) * 100
print("Accuracy : ",accuracy)

features = [[float(i) for i in input(f"Enter {iris.feature_names} : ").split()]]
result = clf.predict(features)

print(iris.target_names[result])
```

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
data,_ = make_blobs(n_samples = 300)

cl = KMeans(n_clusters=3,n_init="auto")
cl.fit(data)

centroids = cl.cluster_centers_
labels = cl.labels_

plt.scatter(data[:,0],data[:,1],c='y')
plt.scatter(centroids[:,0],centroids[:,1],s=100,alpha=0.6,c='black',marker='x')
plt.show()
```

```python
iris = load_iris()
x,y = iris.data,iris.target

x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.7)
cl = KMeans(n_clusters=3,n_init="auto")
cl.fit(x)

centroids = cl.cluster_centers_
labels = cl.labels_

plt.scatter(x[:,0],x[:,1],c=labels,label="Data Points")
plt.scatter(centroids[:,0],centroids[:,1],marker='x',c='red',s=50,label="Centroids")
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
import numpy as np
```

```python
np.random.seed(42)

x = np.random.rand(100,1)
y = 4 + 2 * x + np.random.rand(100,1)

x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.2)
reg = LinearRegression()

reg.fit(x_train,y_train)
pred = reg.predict(x_test)

mse = mean_squared_error(y_test,pred)
r2 = r2_score(y_test,pred)

print(mse,r2)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,c='g',label="Actual Data")
plt.plot(x_test,pred,label="Prediction")
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression as LR
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

x = np.random.rand(100,1)
y = 4 + 2 * x + np.random.rand(100,1)

x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.7)
reg = LR()
reg.fit(x_train,y_train)

y_pred = reg.predict(x_test)

mse = mean_squared_error(y_test,y_pred)
r2  = r2_score(y_test,y_pred)

print(mse,r2)

plt.scatter(x,y,c='g')
plt.plot(x_test,y_pred)
plt.show()

user_input = float(input("Enter a value for x : "))
pred_output = reg.predict(np.array([[user_input]]))

print(pred_output)
```
