```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,6,18])
y = np.array([3,10,12,20])

plt.plot(x,y,color='r',marker='o',markeredgecolor='g',markerfacecolor='g',markersize=10,linestyle='dashed')
plt.show()
```

```python
xpoints = np.array([12,14,16,18,20,22,24])
ypoints = np.array([100,200,250,400,300,450,500])

plt.plot(xpoints,ypoints)
plt.ylabel("Sales")
plt.xlabel("Temperature in degree celsius")
plt.show()
```

```python
x = []
y = []

for line in open("C:/Users/Rahul/Desktop/DS LAB/marks.txt","r"):
    lines = [i for i in line.split()]
    x.append(lines[0])
    y.append(int(lines[1]))

plt.plot(x,y,marker='o',mec='g',mfc='g',c='red',ms=10)
plt.title("Student Marks")
plt.xlabel("Roll No")
plt.ylabel("Marks")
plt.show()
```

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1,y1,c='g',label="line one")
plt.plot(x2,y2,label="line two")
plt.legend()
plt.show()
```

```python
women = np.array([25,32,30,35,29])
men = np.array([22,30,35,35,26])

bar_width = 0.25

plt.bar([1,2,3,4,5],men,width=bar_width)
plt.bar([1.25,2.25,3.25,4.25,5.25],women,width=bar_width)
plt.xticks([1.12,2.12,3.12,4.12,5.12],['G1','G2','G3','G4','G5'])
plt.show()
```

```python
languages = np.array(["Java","Python","PHP","Javascript","C#","C++"])
data = np.array([22.2,17.6,8.8,8,77,6.7])

plt.pie(data,labels=languages)
plt.show()
```

```python
import csv
data = []
label = []

with open("C:/Users/Rahul/Desktop/DS LAB/medals.csv") as file:
    csvFile = csv.reader(file)
    next(csvFile)
    for lines in csvFile:
        label.append(lines[0])
        data.append(lines[1])
plt.pie(data,labels=label)
plt.show()
```

```python
x1=np.array([88, 92, 80, 89, 100, 80, 60, 100, 80, 34])
x2=np.array([35, 79,79, 48, 100, 88, 32, 45, 20, 30])

y = np.arange(10,110,10)

plt.scatter(x1,y,s=50,c='red',label="math marks")
plt.scatter(x2,y,s=50,c='#28d6ed',label="science marks")

plt.title("Scores by person")
plt.xlabel("Marks Range")
plt.ylabel("Marks scored")
plt.legend(bbox_to_anchor=(1.35,1.02))

plt.show()
```
